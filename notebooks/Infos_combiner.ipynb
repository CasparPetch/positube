```python
import os
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir(mypath[0]) if isfile(join(mypath[0], f))]
```

```python
onlyfiles
```

```python
csvs = {}
for x in onlyfiles:
    if x[:5] == 'infos':
        a, b = x.split('.')
        print('loading ', a)
        csvs[a] = pd.read_csv(x, index_col=0, lineterminator='\n')
        print('SUCCESS\n-----')

csvs
```

```python
for k,v in csvs.items():
    print(csvs[k])
```

```python
clean_csvs_list = list(csvs.values())
final_df = pd.concat(clean_csvs_list)
final_df.reset_index(inplace=True, drop=True)
```
